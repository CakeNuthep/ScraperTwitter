```python
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from time import sleep
import csv
```

```python
def getSpandText(articleElement):
  list_span = articleElement.find_elements_by_tag_name("span")
  text = ''
  for span in list_span:
    text += span.text
  return text


def getAllArticle(listArticle):
  listText = []
  for article in listArticle:
    listText.append(getSpandText(article))
  return listText
```

```python
driver.get('https://twitter.com/search?q=%23การเมือง')
# print(driver.page_source) # results
# driver.find_element_by_css_selector(".u-size2of3").click()
articles = []

i = 0
lenOfPage = driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
sleep(10)
articles.extend(getAllArticle(driver.find_elements_by_tag_name("article")))
match = False
while(match == False):
    lastCount = lenOfPage
    
    lenOfPage = driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
    sleep(10)
    element = getAllArticle(driver.find_elements_by_tag_name("article"))
    articles.extend(element)
    if lastCount == lenOfPage:
        match = True

print(len(articles))
```
